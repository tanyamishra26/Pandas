```python
# method='ffill'
#Fills each NaN with the previous non-null value:
students['name'].fillna(method='ffill')
```

```python
# method='bfill' 
# Fills each NaN with the next non-null value:
students['name'].fillna(method='bfill')
```

### **17. apply** (for series and dataframe)

```python
temp.apply(sigmoid)
```

```python
# for dataframe
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
# row wise traverse 
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```
