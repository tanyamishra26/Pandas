## How to fix incomplete data

```python
# create Dummy data

s = {
    'A':[1,2,3],
    'B':[4,np.nan,6],
    'C':[7,8,9],
    'D':[10,11,12]
}
dummy = pd.DataFrame(s)
dummy
```

```python
# drop rows having nan data
dummy.dropna(axis=0)
```

```python
dummy.dropna(axis=1)
```

```python
dummy
```

```python
# drop with limit
dummy.dropna(thresh=2,axis=1)
```

```python
#Replace nan with values

dummy.fillna(value="X")
```

```python
dummy.fillna(value=dummy['B'].mean())
```

```python
dummy.fillna(value=dummy['B'].max())
```

```python
dummy.fillna(value=dummy['B'].sum())
```

## Data Frame Group by

```python
import pandas as pd
import numpy as np
# Create Data
stuff = {
    'Corporation':['Apple', 'Google', 'Meta', 'Apple', 'Google','Meta'],
    'Employees':['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary':[200, 220, 190, 130, 120, 150]}
# Create Dataframe
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Group by corporation - to get object location in memory

company = my_df.groupby('Corporation')
```

```python
company.sum()
company.mean()
company.count()
## max(),min()
## std(),var(),count()
company.describe()
```
